3. Use the Tweepy library to query additional data via the Twitter API (tweet_json.txt)

```python
import tweepy
from tweepy import OAuthHandler
import json
from timeit import default_timer as timer
import numpy as np
```

```python
# Query Twitter API for each tweet in the Twitter archive and save JSON in a text file
# These are hidden to comply with Twitter's API terms and conditions
consumer_key = 'HIDDEN'
consumer_secret = 'HIDDEN'
access_token = 'HIDDEN'
access_secret = 'HIDDEN'
```

```python
auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)
```

```python
api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
# Tweet IDs for which to gather additional data via Twitter's API
tweet_ids = df_tae.tweet_id.values
len(tweet_ids)
```

```python
# Query Twitter's API for JSON data for each tweet ID in the Twitter archive
count = 0
fails_dict = {}
start = timer()
# Save each tweet's returned JSON as a new line in a .txt file
with open('tweet_json.txt', 'w') as outfile:
    # This loop will likely take 20-30 minutes to run because of Twitter's rate limit
    for tweet_id in tweet_ids:
        count += 1
        print(str(count) + ": " + str(tweet_id))
        try:
            tweet = api.get_status(tweet_id, tweet_mode='extended')
            print("Success")
            json.dump(tweet._json, outfile)
            outfile.write('\n')
        except tweepy.TweepyException as e:
            print("Fail")
            fails_dict[tweet_id] = e
            pass
end = timer()
print(end - start)
print(fails_dict)
```

```python
with open('tweet-json.txt') as f:
    lines = f.readlines()
    
data = []
for line in lines:
    data.append(json.loads(line))

data_json = pd.DataFrame(data)
data_json
    
```

To make this comparism,the count of each dog stage has to be put into consideration('None' is not a dog stage, it stands for dogs that there stages were not indicated  . I will define a function that will divide the sum of the ratings of each category by the count of that category.

```python
def average (total, count):
    """ a function that returns average by dividing total of the rating of each dog stage with its count
    """
    return total/count
```

```python
total = tam.groupby('dog_stages').sum()['rating_num'].sort_values() 
count = tam.dog_stages.value_counts()
```

```python
#calculating the average rating for each dog stage
average(total, count).sort_values(ascending = False)
```

```python
# ploting in a pie chart
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

average(total, count).plot(kind = 'pie', autopct = '%.2f%%')
plt.title('average rating for each dog stage');
```
